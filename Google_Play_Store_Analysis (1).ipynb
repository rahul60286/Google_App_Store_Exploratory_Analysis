# Google Play Store Data Analysis

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python

df_store = pd.read_csv('R:/ML PROJECTS/Google app/googleplaystore.csv')
```

# Data Cleaning And Preprocessing

```python
df_store.isnull().sum()
```

```python
df_store.duplicated()
```

```python
# Handling Missing values 
```

```python
# drop rows with missing values
df_store.dropna(inplace=True)
```

```python
# Removes duplicates rows 
df_store.drop_duplicates(inplace=True)
```

```python
df_store.info()
```

```python
df_store.isnull().sum()
```

```python
df_store.duplicated().sum()
```

```python
# Removing Outliers
z_scores = np.abs((df_store['Rating'] - df_store['Rating'].mean()) / df_store['Rating'].std())
df_store = df_store[z_scores < 3]  # Keep rows within 3 standard deviations
```

```python
# change te data type of install colums to numerical data type and removes th comma and the plus symobl
df_store['Installs'] = df_store['Installs'].str.replace('+', '')
df_store['Installs'] = df_store['Installs'].str.replace(',', '')


df_store['Installs'] = df_store['Installs'].astype(int)
```

```python

df_store['Reviews'] = df_store['Reviews'].astype(int)
```

```python
df_store.info()
```

# Data Visualization Techniques

```python
# Bar plot
plt.figure(figsize=(10,6))
sns.countplot(data=df_store, x='Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of App Category')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot

plt.figure(figsize=(10,6))
sns.scatterplot(data=df_store, x='Reviews', y='Rating')
plt.xlabel('Number of Reviews')
plt.ylabel('Rating')
plt.title('Relation between Reviews and Rating')
```

```python
# Box plot

plt.figure(figsize=(10,6))
sns.boxplot(data=df_store, x='Category', y='Installs')
plt.xlabel('Category')
plt.ylabel('Installs')
plt.xticks(rotation =90)
plt.title('Distribution of Installs across Category')
```

# Statistical Analysis And Hypothesis Testing 

```python
df_store.head()
```

```python
from scipy.stats import ttest_ind

# Splitting the dataset into free and paid apps
free_apps = df_store[df_store['Type'] == 'Free']
paid_apps = df_store[df_store['Type'] == 'Paid']
```

```python
# Performing t-test
t_stat, p_value = ttest_ind(free_apps['Rating'], paid_apps['Rating'])
```

```python
p_value
```

```python
# Check the p-value
if p_value < 0.05:
    print("Rejects null hypothesis: there is a significant difference in ratings.")
else:
    print("Fail to rejects null hypothesis: There is no significance difference in ratings ")
```

# Data Storytelling 

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df_store, x='Rating', y= 'Installs', hue='Category')
plt.grid(True)
plt.xlabel('Rating')
plt.ylabel('Number of install')
plt.title('Relations Between Rating and Install')
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.show

```

```python
# Line plot of Average Rating over Time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_store, x='Last Updated', y='Rating')
plt.xlabel('Last Updated')
plt.ylabel('Average Rating')
plt.title('Average Rating Trend over Time')
```

```python
# Convert 'Last Updated' to datetime
df_store['Last Updated'] = pd.to_datetime(df_store['Last Updated'])

# Set 'Last Updated' as the index
df_store.set_index('Last Updated', inplace=True)

# Resample 'Rating' to monthly frequency
df_store_monthly = df_store['Rating'].resample('M').mean()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_store_monthly)
plt.xlabel('Date of Last Update (Monthly)')
plt.xticks(rotation=90)
plt.ylabel('Average Rating')
plt.title('Average Rating Trend over Time (Monthly)')
plt.show()
```
